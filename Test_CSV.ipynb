    56 to 66 ---> Strong majority<br>

</div>

<h5>Percentages of Opposition:</h5>
    <div>0 to 25 ---> Weak opposition<br>
    26 to 40 ---> Medium opposition<br>
    41 to 50 ---> Strong opposition<br>
    Over 50 ---> potential new majority<br>
</div>

<h5>N° Majority Parties:</h5>
    <div> 1 ---> Single Party Majority<br>
    2 to 4 ---> Coalition Majority<br>
    Over 5 ---> Broad Coalition Majority<br>
</div>

<h5>N° Opposition Parties:</h5>
    <div> 1 ---> Single Party Opposition <br>
    2 to 4 ---> Differentiated Opposition<br>
    Over 5 ---> Fragmented Opposition<br>
</div>

<h5>Distribution Index for Majority x100:</h5>
    <div>0 ---> Single party majority<br>
    1 to 10 ---> Nearly one sided majority<br>
    11 to 30 ---> Strong lead majority<br>
    31 to 75 ---> Fragmented majority<br>
    Over 76 ---> Multi-headed majority<br>
</div>

<h5>Distribution Index for Opposition x100:</h5>
    <div>0 to 20 ---> Cohesive Opposition<br>
    21 to 45 ---> Diverse Opposition<br>
    46 to 75 ---> Fragmented Opposition<br>
    Over 76 ---> Higly Fragmented/ Multi-Headed Opposition<br>
</div>

<h5>Distribution Index for Government Seats x100:</h5>
    <div>0 ---> Monochromatic Government<br>
    1 to 25 ---> Governement with Leading Party<br>

    56 to 75 ---> Large Coalition Government<br>
    Over 75 ---> Super Coalition/National Unity Government<br>
</div>

<h5>Members of the Government</h5>
<div> Under 22 ---> Small Government<br>
    23 to 28 ---> Medium Government<br>

</div>

<h5>Swapped Ministers</h5>

```python
#Preparing Government duration columns
duration_labels = ['U','WS','AS','SS','HS']
duration_intervals = [0,100,250,450,1000, 10000]
df['Time in charge (days)'] = pd.DataFrame(pd.cut(df['Time in charge (days)'], duration_intervals, labels=duration_labels))
df['Effective Days'] = pd.DataFrame(pd.cut(df['Effective Days'], duration_intervals, labels=duration_labels))
df.head()
```

```python
#Preparing Majority and Opposition columns

percentages = []
for i in range(len(df)):
    tot_seats = df['House Majority'][i] + df['House Opposition'][i] + df['Senate Majority'][i] + df['Senate Opposition'][i] 
    perc_m = int(round((df['House Majority'].iloc[i]+df['Senate Majority'].iloc[i])/tot_seats,2)*100)
    perc_o = int(round((df['House Opposition'].iloc[i]+df['Senate Opposition'].iloc[i])/tot_seats,2)*100)
    percentages.append([perc_m, perc_o])

perc_df = pd.DataFrame(percentages, columns=['Tot % Majority','Tot % Opposition'])

majority_intervals = [0,50,55,66,100]
majority_labels = ['NM', 'WM', 'SM', 'CM']
opposition_intervals = [0,25,40,50,100]
opposition_labels = ['WO', 'MO', 'SO', 'OM']

perc_df['Tot % Majority'] = pd.DataFrame(pd.cut(perc_df['Tot % Majority'], majority_intervals, labels=majority_labels))
perc_df['Tot % Opposition'] = pd.DataFrame(pd.cut(perc_df['Tot % Opposition'], opposition_intervals, labels=opposition_labels))

df = pd.concat([df.drop(['House Majority', 'House Opposition', 'Senate Majority', 'Senate Opposition'], axis=1), perc_df], axis=1)

df.head()
```

```python
print(df.columns)
```

```python
#Preparing Distribution Indexes columns
dis = []
for i in range(len(df)):
    di_m = int(round((df['DI House Majority'].iloc[i]+df['DI Senate Majority'].iloc[i])/2,2)*100)
    di_o = int(round((df['DI House Opposition'].iloc[i]+df['DI Senate Opposition'].iloc[i])/2,2)*100)
    dis.append([di_m, di_o])

di_df = pd.DataFrame(dis, columns=['Tot DI Majority','Tot DI Opposition'])
di_majority_intervals = [-1,0,10,30,75,100]
di_majority_labels = ['SP', 'NO', 'SL', 'F', 'MH']
di_opposition_intervals = [0,20,45,75,100]
di_opposition_labels = ['CO', 'DO', 'SO', 'MH']

di_df['Tot DI Majority'] = pd.DataFrame(pd.cut(di_df['Tot DI Majority'], di_majority_intervals, labels=di_majority_labels))
di_df['Tot DI Opposition'] = pd.DataFrame(pd.cut(di_df['Tot DI Opposition'], di_opposition_intervals, labels=di_opposition_labels))

df = pd.concat([df.drop(['DI House Majority', 'DI House Opposition', 'DI Senate Majority', 'DI Senate Opposition'], axis=1), di_df], axis=1)
df.head()
```

```python
#Preparing Government Members Columns
df['DI Parties in Government'] = 100*df['DI Parties in Government']
df['DI Parties in Government'] = df['DI Parties in Government'].astype(int)
df['Tot Gov. Members'] = df['Tot Gov. Members'].astype(int)
df = df.drop(['N° of VP', 'N° Ministers', 'N° Ministers without portfolio'], axis=1)

di_gov_intervals = [-1,1,25,55,75,100]
di_gov_labels = ['M', 'L', 'C', 'LC', 'SC']
gov_m_intervals = [0,22,28,100]
gov_m_labels = ['S', 'M', 'L']
swap_min_intervals = [-1,0,3,6,20]
swap_min_labels = ['U', 'S', 'M','L']

df['DI Parties in Government'] = pd.DataFrame(pd.cut(df['DI Parties in Government'], di_gov_intervals, labels=di_gov_labels))
df['Tot Gov. Members'] = pd.DataFrame(pd.cut(df['Tot Gov. Members'], gov_m_intervals, labels=gov_m_labels))
df['Swapped Ministers'] = pd.DataFrame(pd.cut(df['Swapped Ministers'], swap_min_intervals, labels=swap_min_labels))
df.head()
```

```python
#Replacing all Boolean Columns with 0 and 1
df = df.replace({True:1, False:0})
df['Fell Unnaturally'] = df['Fell Unnaturally'].fillna(1).astype(int)
df.head()
```

```python
#Preparing Parties Number Columns
npm_intervals = [0,1,4,15]
npm_labels = ['S', 'C', 'BC']
npo_intervals = [0,1,4,15]
npo_labels = ['S', 'D', 'F']

df['N° Majority Parties'] = pd.DataFrame(pd.cut(df['N° Majority Parties'].astype(int), npm_intervals, labels=npm_labels))
df['N° Opposition Parties'] = pd.DataFrame(pd.cut(df['N° Opposition Parties'].astype(int), npo_intervals, labels=npo_labels))

df.head()
```

```python
#Encoding Years and Nominals
years_intervals = [0,1950, 1960,1970,1980, 1990,2000,2010,2020,2030]
years_labels = ['40s', '50s', '60s', '70s', '80s', '90s', '2000s', '2010s', '2020s']
legislation_intervals = [0,12,50]
legislation_labels = ['1R','2R']
nominal_replacements = {
    'External Pressure':'EP',
    'Internal Struggle':'IS',
    'End of Legislation':'EL',
    'No confidence':'NC',
    'Lost Confidence':'LC',
    'Reshuffle':'R'
}


df['First Year'] = pd.DataFrame(pd.cut(df['First Year'].astype(int), years_intervals, labels=years_labels))
df['Legislation'] = pd.DataFrame(pd.cut(df['Legislation'].astype(int), legislation_intervals, labels=legislation_labels))
df = df.replace(nominal_replacements)


df.head()
```
